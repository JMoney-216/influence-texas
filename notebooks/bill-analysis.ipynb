```python
from bs4 import BeautifulSoup

senate_doc = "./data/analysis/HB00004S.htm"
house_doc = "./data/analysis/HB00053H.htm"

soup = BeautifulSoup(open(senate_doc), 'html.parser')

print(soup.prettify())
```

```python
commonHeadings = ["BACKGROUND AND PURPOSE", "AUTHOR'S /\nSPONSOR'S STATEMENT OF INTENT"]

def getHeading(soup):
    
    heading = soup.find("u", string = commonHeadings)
    if heading == None:
        try:
            if soup.find("h1").string == "BILL ANALYSIS":
                heading = soup.find_all("u")[0]
            else:
                heading = soup.find_all("u")[1]
        except:
            heading = soup.find_all("u")[1]
    return heading
```

```python
def getAnalysis(heading):
    response = ""
    for element in heading.next_elements:
        try:
            if element.string == 'RULEMAKING\nAUTHORITY' or \
            element.string == "CRIMINAL JUSTICE IMPACT":
                return response.replace("\n", " ").replace("\xa0", "").strip()
        except:
            pass
        try:
            if element.name == "u":
                return response
            elif element.string != None and element.string not in commonHeadings:
                if response.endswith(element.string) == False:
                    response = response + element.string + "\n"
        except: 
            response = response + element.string
    return response.replace("\n", " ").replace("\xa0", "").strip()

getAnalysis(heading)
```

```python
def getAnalysis(heading):
    response = ""
    for element in heading.parent.parent.next_siblings:
        if element.string == "RULEMAKING\nAUTHORITY":
            return response.strip()
        if element.string != None:
            response = response + element.string
    return response.strip()
```
